#### Monte-Carlo

###### Option Pricing
Monte Carlo methods excel at pricing complex derivatives where analytical solutions don't exist. For European options, you simulate thousands of possible stock price paths using geometric Brownian motion, then calculate the average payoff discounted to present value. For exotic options like Asian options (average price) or barrier options, Monte Carlo often provides the only practical pricing method.

###### Risk Management
Value at Risk (VaR) calculations rely heavily on Monte Carlo simulation. By generating numerous scenarios for portfolio returns based on historical correlations and volatilities, firms can estimate potential losses at specific confidence levels. This helps determine capital requirements and position limits. 

###### Portfolio Optimization
When optimizing portfolios with non-normal return distributions or complex constraints, Monte Carlo simulation can explore the entire solution space. It's particularly useful for incorporating transaction costs, liquidity constraints, and regime changes that analytical methods struggle to handle. 

###### Credit Risk Modeling 
Banks use Monte Carlo methods to model default correlations and estimate portfolio credit losses. The simulation generates correlated default scenarios across borrowers, accounting for economic cycles and sector concentrations that drive simultaneous defaults.

###### Interest Rate and Fixed Income
Modeling interest rate movements requires simulating entire yield curves over time. Monte Carlo methods can incorporate mean reversion, volatility clustering, and term structure dynamics to price bonds, mortgage-backed securities, and interest rate derivatives.

###### Stress Testing
Regulatory stress tests often employ Monte Carlo simulation to generate adverse economic scenarios. Banks simulate how their portfolios would perform under various recession scenarios, helping regulators assess systemic risk.

#### MonteCarlo code

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Define simulation parameters
annual_contribution = 6000           # Roth IRA yearly contribution
years = 40                           # Investment duration
mean_return = 0.07                   # Expected average annual return (7%)
std_dev = 0.15                       # Standard deviation of returns (15%)
num_simulations = 10000             # Number of Monte Carlo simulations

# Step 2: Run Monte Carlo simulation
final_portfolio_values = []         # To store the final value of each simulation

for _ in range(num_simulations):
    portfolio_value = 0
    for _ in range(years):
        annual_return = np.random.normal(loc=mean_return, scale=std_dev)
        portfolio_value = (portfolio_value + annual_contribution) * (1 + annual_return)
    final_portfolio_values.append(portfolio_value)

# Convert to numpy array for easier analysis
final_portfolio_values = np.array(final_portfolio_values)

# Step 3: Summary statistics
mean_val = np.mean(final_portfolio_values)
median_val = np.median(final_portfolio_values)
std_val = np.std(final_portfolio_values)
percentiles = np.percentile(final_portfolio_values, [5, 25, 75, 95])

# Step 4: Plot histogram
plt.figure(figsize=(12, 6))
plt.hist(final_portfolio_values, bins=100, color='skyblue', edgecolor='black')
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean: ${mean_val:,.0f}')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label=f'Median: ${median_val:,.0f}')
plt.title("Monte Carlo Simulation of 40-Year Roth IRA Investment")
plt.xlabel("Portfolio Value at Retirement ($)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Show summary
{
    "Mean Portfolio Value": f"${mean_val:,.2f}",
    "Median Portfolio Value": f"${median_val:,.2f}",
    "Standard Deviation": f"${std_val:,.2f}",
    "5th Percentile": f"${percentiles[0]:,.2f}",
    "25th Percentile": f"${percentiles[1]:,.2f}",
    "75th Percentile": f"${percentiles[2]:,.2f}",
    "95th Percentile": f"${percentiles[3]:,.2f}",
}
```

#### Formula for Yearly Growth with Contribution:
At each year, the portfolio grows like this:

\begin{equation*}
V_{\text {year }}=\left(V_{\text {previous }}+C\right) \times(1+r)
\end{equation*}


Where:
- $V_{\text {previous }}$ is the portfolio value at the end of the previous year
- $C$ is the annual contribution (e.g. $\$ 6,000$ )
- $r$ is a random return drawn from a normal distribution:

#### For a particular stock

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
from scipy import stats
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Step 1: Download Historical Data
print("Step 1: Downloading historical stock data...")
end = dt.datetime.now()
start = end - dt.timedelta(days=365 * 5)  # 5 years of data
assets = ["SPY"]
df = yf.download(assets, start=start, end=end, auto_adjust=False)["Adj Close"]

print(f"Downloaded {len(df)} days of data for {assets[0]}")
print(f"Date range: {df.index[0].date()} to {df.index[-1].date()}")


print("="*100)
# Step 2: Calculate Daily Returns
daily_returns = df.pct_change().dropna()

# Calculate key statistics
mean_daily_return = daily_returns.mean()
std_daily_return = daily_returns.std()

# Convert to annual figures (252 trading days per year)
# Extract scalar values from pandas Series
annual_return = float(mean_daily_return * 252)
annual_volatility = float(std_daily_return * np.sqrt(252))

print(f"Historical annual return: {annual_return:.2%}")
print(f"Historical annual volatility: {annual_volatility:.2%}")



print("="*100)
# Step 3: Set Simulation Parameters
initial_investment = 5000  # Initial lump sum
annual_contribution = 6000  # Fixed annual contribution ($1000/month)
years = 40  # Investment horizon
num_simulations = 10000  # Number of Monte Carlo runs

print(f"Initial investment: ${initial_investment:,}")
print(f"Annual contribution: ${annual_contribution:,}")
print(f"Investment period: {years} years")
print(f"Number of simulations: {num_simulations:,}")

# Step 4: Monte Carlo Simulation Function
def monte_carlo_portfolio_simulation(initial_investment, annual_contribution, 
                                   annual_return, annual_volatility, 
                                   years, num_simulations):
    
    # Initialize arrays to store results
    final_values = np.zeros(num_simulations)
    all_paths = np.zeros((num_simulations, years + 1))
    
    for i in range(num_simulations):
        # Start with initial investment
        portfolio_value = initial_investment
        path = [portfolio_value]
        
        for year in range(years):
            # Generate random annual return using normal distribution
            random_return = np.random.normal(annual_return, annual_volatility)
            
            # Portfolio Growth Formula: Portfolio_Value(t+1) = Portfolio_Value(t) × (1 + Return) + Annual_Contribution
            portfolio_value = portfolio_value * (1 + random_return) + annual_contribution
            
            # Store the path
            path.append(portfolio_value)
        
        # Store final value and complete path
        final_values[i] = portfolio_value
        all_paths[i] = path
    
    return final_values, all_paths
```

```python
# Step 5: Run the Simulation

print("Running Monte Carlo simulation...")
print("="*100)

final_portfolio_values, portfolio_paths = monte_carlo_portfolio_simulation(
    initial_investment, annual_contribution, annual_return, annual_volatility, 
    years, num_simulations
)


# Step 6: Calculate Statistics
print("\nStep 6: Analyzing results...")
print("="*100)

# Calculate total contributions
total_contributions = initial_investment + (annual_contribution * years)

# Calculate statistics
mean_final_value = np.mean(final_portfolio_values)
median_final_value = np.median(final_portfolio_values)
std_final_value = np.std(final_portfolio_values)

# Calculate percentiles
percentiles = [5, 10, 25, 50, 75, 90, 95]
portfolio_percentiles = np.percentile(final_portfolio_values, percentiles)

# Calculate probability of loss
prob_loss = np.mean(final_portfolio_values < total_contributions) * 100

print(f"\nTotal amount invested: ${total_contributions:,.0f}")
print(f"Mean final portfolio value: ${mean_final_value:,.0f}")
print(f"Median final portfolio value: ${median_final_value:,.0f}")
print(f"Standard deviation: ${std_final_value:,.0f}")
print(f"Probability of loss: {prob_loss:.1f}%")

print(f"\nPercentile Analysis:")
for i, p in enumerate(percentiles):
    print(f"{p}th percentile: ${portfolio_percentiles[i]:,.0f}")
```

```python
# Step 7: Visualization

# Create subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Monte Carlo Portfolio Simulation Results', fontsize=16, fontweight='bold')

# Plot 1: Sample portfolio paths
ax1.set_title('Sample Portfolio Growth Paths (100 simulations)')
years_axis = range(years + 1)
for i in range(min(100, num_simulations)):
    ax1.plot(years_axis, portfolio_paths[i], alpha=0.1, color='blue')

# Add mean path
mean_path = np.mean(portfolio_paths, axis=0)
ax1.plot(years_axis, mean_path, color='red', linewidth=3, label='Mean Path')
ax1.axhline(y=total_contributions, color='green', linestyle='--', 
           label=f'Total Contributions: ${total_contributions:,.0f}')
ax1.set_xlabel('Years')
ax1.set_ylabel('Portfolio Value ($)')
ax1.set_xlim(0, years)  # X-axis from 0 to 40 years
ax1.set_ylim(0, 100000000)  # Y-axis from 0 to max portfolio value + 10% buffer
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# Plot 2: Distribution of final values
ax2.set_title('Distribution of Final Portfolio Values')
ax2.hist(final_portfolio_values, bins=100, alpha=0.7, color='skyblue', edgecolor='black')
ax2.axvline(mean_final_value, color='red', linestyle='-', linewidth=2, 
           label=f'Mean: ${mean_final_value:,.0f}')
ax2.axvline(median_final_value, color='orange', linestyle='-', linewidth=2, 
           label=f'Median: ${median_final_value:,.0f}')
ax2.axvline(total_contributions, color='green', linestyle='--', linewidth=2, 
           label=f'Total Invested: ${total_contributions:,.0f}')
ax2.set_xlabel('Final Portfolio Value ($)')
ax2.set_ylabel('Frequency')
#ax2.set_xlim(-1000, 500000000)
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000000:.1f}M'))

# Plot 3: Percentile chart
ax3.set_title('Portfolio Value Percentiles')
colors = plt.cm.RdYlGn(np.linspace(0.2, 0.8, len(percentiles)))
bars = ax3.bar(range(len(percentiles)), portfolio_percentiles, color=colors)
ax3.set_xlabel('Percentile')
ax3.set_ylabel('Portfolio Value ($)')
ax3.set_xticks(range(len(percentiles)))
ax3.set_xticklabels([f'{p}th' for p in percentiles])
ax3.grid(True, alpha=0.3)
ax3.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000000:.1f}M'))

# Add value labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
             f'${height/1000000:.1f}M', ha='center', va='bottom', fontsize=9)

# Plot 4: Risk-Return Analysis over time
ax4.set_title('Portfolio Value Confidence Intervals Over Time')
percentile_paths_5 = np.percentile(portfolio_paths, 5, axis=0)
percentile_paths_25 = np.percentile(portfolio_paths, 25, axis=0)
percentile_paths_75 = np.percentile(portfolio_paths, 75, axis=0)
percentile_paths_95 = np.percentile(portfolio_paths, 95, axis=0)

ax4.fill_between(years_axis, percentile_paths_5, percentile_paths_95, 
                alpha=0.2, color='blue', label='90% Confidence Interval')
ax4.fill_between(years_axis, percentile_paths_25, percentile_paths_75, 
                alpha=0.4, color='blue', label='50% Confidence Interval')
ax4.plot(years_axis, mean_path, color='red', linewidth=2, label='Mean Path')
ax4.set_xlabel('Years')
ax4.set_ylabel('Portfolio Value ($)')
ax4.legend()
ax4.grid(True, alpha=0.3)
ax4.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

plt.tight_layout()
plt.savefig('portfolio_simulation.png', 
           dpi=300,                    # High resolution
           bbox_inches='tight',        # Remove extra whitespace
           facecolor='white',          # Background color
           edgecolor='none')  
plt.show()

```
