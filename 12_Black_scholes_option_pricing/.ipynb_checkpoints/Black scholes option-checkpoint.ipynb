### Theory

#### Black-Scholes Model

The Black-Scholes model is a mathematical model used to price European call and put options. It gives a closed-form formula based on certain assumptions.
It was developed by Fischer Black, Myron Scholes, and Robert Merton in the early 1970s and revolutionized options trading.

##### Assumptions
- The option is European (can be exercised only at expiration).
- No dividends are paid.
- Markets are efficient.
- No transaction costs.
- The risk-free interest rate and volatility are constant.
- The stock price follows a log-normal distribution (i.e., geometric Brownian motion).

##### The Formula

For a Call Option:

\begin{equation*}
C=S \cdot N\left(d_1\right)-K \cdot e^{-r T} \cdot N\left(d_2\right)
\end{equation*}


For a Put Option:

\begin{equation*}
P=K \cdot e^{-r T} \cdot N\left(-d_2\right)-S \cdot N\left(-d_1\right)
\end{equation*}


Where:
- $S=$ Current stock price
- $K=$ Strike price
- $T=$ Time to maturity (in years)
- $r=$ Risk-free interest rate
- $\sigma=$ Volatility of the stock (standard deviation)
- $N()=$ Cumulative distribution function of the standard normal distribution

\begin{equation}
d_1=\frac{\ln (S / K)+\left(r+\sigma^2 / 2\right) T}{\sigma \sqrt{T}}, \quad d_2=d_1-\sigma \sqrt{T}
\end{equation}

### Code-1

```python
import math
from scipy.stats import norm
import matplotlib.pyplot as plt

# Define option parameters
S = 100    # Current stock price
K = 100    # Strike price
T = 1      # Time to maturity (1 year)
r = 0.05   # Risk-free interest rate (5%)
sigma = 0.2  # Volatility (20%)

# Calculate d1 and d2
d1 = (math.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * math.sqrt(T))
d2 = d1 - sigma * math.sqrt(T)

# Calculate the Call and Put option prices
call_price = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
put_price = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)

print("\nOption Prices:")
print(f"Call Option Price: {call_price}")
print(f"Put Option Price: {put_price}")
```

```python
# Create a range of stock prices
stock_prices = [i for i in range(50, 151)]  # From 50 to 150

# Calculate call and put prices for each stock price
call_prices = []
put_prices = []

for S_temp in stock_prices:
    d1 = (math.log(S_temp / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)
    call = S_temp * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
    put = K * math.exp(-r * T) * norm.cdf(-d2) - S_temp * norm.cdf(-d1)
    call_prices.append(call)
    put_prices.append(put)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, call_prices, label='Call Option Price', color='blue')
plt.plot(stock_prices, put_prices, label='Put Option Price', color='red')
plt.axvline(x=K, linestyle='--', color='gray', label='Strike Price')
plt.title('Black-Scholes Option Pricing')
plt.xlabel('Stock Price')
plt.ylabel('Option Price')
plt.legend()
plt.grid(True)
plt.savefig("p121.png",dpi=300)
plt.show()
```

### Test

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
import math

# Let's use Apple stock as an example
S0 = 100.0      # Current stock price ($150)
K = 100.0       # Strike price ($155)
T = 1        # Time to expiration (3 months = 0.25 years)
r = 0.05        # Risk-free rate (5% annual)
sigma = 0.2     # Volatility (30% annual)

def black_scholes_call(S, K, T, r, sigma):
    # Calculate d1 and d2
    d1 = (math.log(S/K) + (r + 0.5 * sigma**2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)
    
    # Calculate option price
    call_price = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
    put_price = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    
    return call_price, put_price

# Test our function
call_price, put_price = black_scholes_call(S0, K, T, r, sigma)

print(f"Call price: ${call_price:.2f}")
print(f"Put price: ${put_price:.2f}")
```

```python
# Create plots to visualize the relationships

# Plot 1: Option price vs Stock price
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

stock_range = np.linspace(50, 180, 100)
option_call_prices  = [black_scholes_call(S, K, T, r, sigma)[0] for S in stock_range]
option_put_prices  = [black_scholes_call(S, K, T, r, sigma)[1] for S in stock_range]
intrinsic_call_values = [max(0, S - K) for S in stock_range]
intrinsic_put_values = [max(0, K - S) for S in stock_range]

ax1.plot(stock_range, option_call_prices, 'b-', linewidth=2, label='Black-Scholes Call Price')
ax1.plot(stock_range, option_put_prices, 'g-', linewidth=2, label='Black-Scholes Put Price')
ax1.plot(stock_range, intrinsic_call_values, 'r--', linewidth=2, label='Intrinsic call Value')
ax1.plot(stock_range, intrinsic_put_values, 'y--', linewidth=2, label='Intrinsic put Value')
ax1.axvline(x=K, color='gray', linestyle=':', alpha=0.7, label=f'Strike Price (${K})')
ax1.axvline(x=S0, color='green', linestyle=':', alpha=0.7, label=f'Current Price (${S0})')
ax1.set_xlabel('Stock Price ($)')
ax1.set_ylabel('Option Price ($)')
ax1.set_title('Option Price vs Stock Price')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Option price vs Volatility
vol_range = np.linspace(0.1, 0.8, 30)
vol_call_prices = [black_scholes_call(S0, K, T, r, vol)[0] for vol in vol_range]
vol_put_prices = [black_scholes_call(S0, K, T, r, vol)[1] for vol in vol_range]

ax2.plot(vol_range * 100, vol_call_prices, 'purple', linewidth=2,label='Black-Scholes Call Price')
ax2.plot(vol_range * 100, vol_put_prices, 'red', linewidth=2,label='Black-Scholes Put Price')
ax2.axvline(x=sigma*100, color='red', linestyle='--', alpha=0.7, label=f'Current Vol ({sigma*100}%)')
ax2.set_xlabel('Volatility (%)')
ax2.set_ylabel('Option Price ($)')
ax2.set_title('Option Price vs Volatility')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Option price vs Time to expiration
time_range = np.linspace(0.01, 1, 50)  # 0.01 to 1 year
time_call_prices = [black_scholes_call(S0, K, t, r, sigma)[0] for t in time_range]
time_put_prices = [black_scholes_call(S0, K, t, r, sigma)[1] for t in time_range]

ax3.plot(time_range * 365, time_call_prices, 'orange', linewidth=2,label='Call Price')
ax3.plot(time_range * 365, time_put_prices, 'blue', linewidth=2,label='Put Price')
ax3.axvline(x=T*365, color='red', linestyle='--', alpha=0.7, label=f'Current Time ({T*365:.0f} days)')
ax3.set_xlabel('Days to Expiration')
ax3.set_ylabel('Option Price ($)')
ax3.set_title('Call Option Price vs Time to Expiration')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: Option Greeks (Delta)
# Delta shows how much option price changes with stock price
delta_values = []
for S in stock_range:
    d1 = (math.log(S/K) + (r + 0.5 * sigma**2) * T) / (sigma * math.sqrt(T))
    delta = norm.cdf(d1)
    delta_values.append(delta)

ax4.plot(stock_range, delta_values, 'darkgreen', linewidth=2)
ax4.axvline(x=K, color='gray', linestyle=':', alpha=0.7, label=f'Strike Price (${K})')
ax4.axvline(x=S0, color='red', linestyle='--', alpha=0.7, label=f'Current Price (${S0})')
ax4.set_xlabel('Stock Price ($)')
ax4.set_ylabel('Delta')
ax4.set_title('Option Delta vs Stock Price')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("p121.png",dpi=300)
plt.show()
```
