```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import datetime as dt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
years = 15
endDate = dt.datetime.now()
startDate = endDate - dt. timedelta(days = 365*years)
assets = ["VTI", "SPY", "QQQ", "BND", "GLD"]
df = yf.download(assets, start=startDate, end=endDate, auto_adjust=False)["Adj Close"]
print(f"\nData Shape: {df.shape}")
# Display first and last few rows
print("\nFirst 5 rows:")
print(df.head())
print("\nLast 5 rows:")
print(df.tail())
```

```python
log_returns = np.log(df/df.shift(1)).dropna()
print(log_returns)
```

```python
portfolio_value = 1000000
weights = np.array([1/len(assets)]*len(assets))
print(weights)
```

```python
historical_returns = (log_returns * weights).sum(axis =1)
print(historical_returns)
```

```python
days = 5
range_returns = historical_returns.rolling(window = days).sum()
range_returns = range_returns.dropna()
print(range_returns)
```

```python
confidence_interval = 0.95
VaR = -np.percentile(range_returns, 100 - (confidence_interval * 100))*portfolio_value
print(VaR)
```

```python
return_window = days
range_returns = historical_returns.rolling(window=return_window).sum()
range_returns = range_returns.dropna()
range_returns_dollar = range_returns * portfolio_value


plt.hist(range_returns_dollar.dropna(), bins=50, density=True)
plt.xlabel(f'(return_window)-Day Portfolio Return (Dollar Value)')
plt.ylabel('Frequency')
plt.title(f'Distribution of Portfolio (return_window)-Day Returns (Dollar Value)')
plt.axvline(-VaR, color='b', linestyle='dashed', linewidth=2, label=f'VaR at {confidence_interval:.0%} confldence level')
plt.legend()
plt.show()
```

### Try-2

```python
import datetime as dt
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
years = 15
endDate = dt.datetime.now()
startDate = endDate - dt. timedelta(days = 365*years)
assets = ["VTI", "SPY", "QQQ", "BND", "GLD"]
df = yf.download(assets, start=startDate, end=endDate, auto_adjust=False)["Adj Close"]
print(f"\nData Shape: {df.shape}")
# Display first and last few rows
print("\nFirst 5 rows:")
print(df.head())
print("\nLast 5 rows:")
print(df.tail())
```

```python
# Plot normalized prices to compare performance
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))

# Absolute prices
df.plot(ax=ax1, linewidth=2)
ax1.set_title('Historical Prices - Absolute Values', fontsize=14, fontweight='bold')
ax1.set_ylabel('Price ($)')
ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax1.grid(True, alpha=0.3)

# Normalized prices (base = 100)
normalized_prices = (df / df.iloc[0] * 100)
normalized_prices.plot(ax=ax2, linewidth=2)
ax2.set_title('Historical Prices - Normalized (Base = 100)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Normalized Price')
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Calculate daily returns
returns = df.pct_change().dropna()

print(f"Returns Shape: {returns.shape}")
print("\nDaily Returns Summary Statistics:")
print(returns.describe())

print("\nFirst 5 rows of daily returns:")
print(returns.head())
```

```python
rolling_returns= returns.cumsum()
fig, axes = plt.subplots(3, 5, figsize=(20, 10))  # 3 rows, 5 columns

for i, asset in enumerate(assets):
    returns[asset].plot(ax=axes[0,i], alpha=0.7, color=f'C{i}')
    axes[0,i].set_title(f'{asset} Daily Returns', fontweight='bold')
    axes[0,i].axhline(y=0, color='red', linestyle='--', alpha=0.5)
    axes[0,i].grid(True, alpha=0.3)

    # Row 1: Daily Returns
    rolling_returns[asset].plot(ax=axes[1, i], color=f'C{i}', alpha=0.7)
    axes[1, i].set_title(f'{asset} - Rolling Returns', fontweight='bold')
    axes[1, i].axhline(y=0, color='red', linestyle='--', alpha=0.5)
    axes[1, i].grid(True, alpha=0.3)

    # Row 2: Histogram of Returns
    returns[asset].plot.hist(ax=axes[2, i], bins=50, alpha=0.7, color=f'C{i}')
    axes[2, i].set_title(f'{asset} - Histogram', fontweight='bold')
    axes[2,i].axvline(returns[asset].mean(), color='red', linestyle='--', label=f'Mean: {returns[asset].mean():.4f}')
    axes[2,i].axvline(returns[asset].std(), color='orange', linestyle='--', label=f'Std: {returns[asset].std():.4f}')
    axes[2,i].legend()
    axes[2, i].grid(True, alpha=0.3)

# Adjust layout
plt.tight_layout()
plt.savefig("p91.png",dpi=300)
plt.show()
```

```python
correlation_matrix = returns.corr()
print(correlation_matrix)

# Plot correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5, cbar_kws={"shrink": .8})
plt.title('Asset Returns Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Define portfolio weights (equal weight for simplicity)
weights = np.array([1/len(assets)]*len(assets))  # Equal weights
print("Portfolio Weights:")
for i, asset in enumerate(assets):
    print(f"{asset}: {weights[i]:.1%}")

# Calculate portfolio returns
portfolio_returns = (returns * weights).sum(axis=1)
print(portfolio_returns)
```

```python
print(f"\nPortfolio Returns Shape: {portfolio_returns.shape}")
print("\nPortfolio Returns Summary:")
print(f"Mean Daily Return: {portfolio_returns.mean():.4f} ({portfolio_returns.mean()*252:.2%} annualized)")
print(f"Daily Volatility: {portfolio_returns.std():.4f} ({portfolio_returns.std()*np.sqrt(252):.2%} annualized)")
print(f"Sharpe Ratio (assuming 0% risk-free rate): {portfolio_returns.mean()/portfolio_returns.std()*np.sqrt(252):.2f}")
```

```python

# Portfolio vs individual assets performance
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(18, 12))

# Portfolio returns time series
portfolio_returns.plot(ax=ax1, color='darkblue', alpha=0.7)
ax1.set_title('Portfolio Daily Returns', fontweight='bold')
ax1.axhline(y=0, color='red', linestyle='--', alpha=0.5)
ax1.grid(True, alpha=0.3)

# Portfolio returns distribution
portfolio_returns.hist(bins=50, ax=ax2, alpha=0.7, color='darkblue', edgecolor='black')
ax2.set_title('Portfolio Returns Distribution', fontweight='bold')
ax2.axvline(portfolio_returns.mean(), color='red', linestyle='--', 
           label=f'Mean: {portfolio_returns.mean():.4f}')
ax2.axvline(portfolio_returns.std(), color='orange', linestyle='--', 
           label=f'Std: {portfolio_returns.std():.4f}')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Cumulative returns comparison
cumulative_returns = (1 + returns).cumprod()
portfolio_cumulative = (1 + portfolio_returns).cumprod()

cumulative_returns.plot(ax=ax3, alpha=0.7)
portfolio_cumulative.plot(ax=ax3, color='black', linewidth=3, label='Portfolio')
ax3.set_title('Cumulative Returns Comparison', fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Rolling volatility (30-day)
rolling_vol = portfolio_returns.rolling(window=30).std() * np.sqrt(252)
rolling_vol.plot(ax=ax4, color='red', alpha=0.7)
ax4.set_title('Portfolio Rolling Volatility (30-day, Annualized)', fontweight='bold')
ax4.set_ylabel('Volatility')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Define confidence levels
portfolio_value=1000000
confidence_levels = [0.90, 0.95, 0.99]
days = 5
range_returns = portfolio_returns.rolling(window = days).sum().dropna()
print(range_returns)

var_results = {}
for conf_level in confidence_levels:
    var_amount = -np.percentile(range_returns, (1 - conf_level) * 100)*portfolio_value
    var_results[conf_level] = var_amount
    print(f"{conf_level:.0%} Confidence Level: {var_amount:.4f} ")
```

```python
es_results = {}
for conf_level in confidence_levels:
    var_threshold = var_results[conf_level]
    es = portfolio_returns[portfolio_returns <= var_threshold].mean()
    es_results[conf_level] = es
    print(f"{conf_level:.0%} Confidence Level: {es:.4f} ({es:.2%})")
```

```python
# STEP 7: VALUE AT RISK CALCULATION
print("\n" + "="*50)
print("STEP 7: HISTORICAL VALUE AT RISK CALCULATION")
print("="*50)

# Sort portfolio returns for percentile calculation
sorted_returns = portfolio_returns.sort_values()

# Define confidence levels
confidence_levels = [0.90, 0.95, 0.99]
print("VaR Calculation using Historical Method:")
print("\nHistorical VaR (1-day holding period):")

var_results = {}
for conf_level in confidence_levels:
    var_percentile = np.percentile(portfolio_returns, (1 - conf_level) * 100)
    var_results[conf_level] = var_percentile
    print(f"{conf_level:.0%} Confidence Level: {var_percentile:.4f} ({var_percentile:.2%})")

# Calculate Expected Shortfall (Conditional VaR)
print("\nExpected Shortfall (Conditional VaR):")
es_results = {}
for conf_level in confidence_levels:
    var_threshold = var_results[conf_level]
    es = portfolio_returns[portfolio_returns <= var_threshold].mean()
    es_results[conf_level] = es
    print(f"{conf_level:.0%} Confidence Level: {es:.4f} ({es:.2%})")
```

```python
# Create comprehensive VaR visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(18, 12))

# Histogram with VaR levels
portfolio_returns.hist(bins=50, ax=ax1, alpha=0.7, color='skyblue', edgecolor='black')
ax1.set_title('Portfolio Returns Distribution with VaR Levels', fontweight='bold')

colors = ['orange', 'red', 'darkred']
for i, conf_level in enumerate(confidence_levels):
    ax1.axvline(var_results[conf_level], color=colors[i], linestyle='--', linewidth=2,
               label=f'VaR {conf_level:.0%}: {var_results[conf_level]:.3f}')
    ax1.axvline(es_results[conf_level], color=colors[i], linestyle=':', linewidth=2,
               label=f'ES {conf_level:.0%}: {es_results[conf_level]:.3f}')

ax1.legend()
ax1.grid(True, alpha=0.3)

# Q-Q plot to check normality
stats.probplot(portfolio_returns, dist="norm", plot=ax2)
ax2.set_title('Q-Q Plot: Portfolio Returns vs Normal Distribution', fontweight='bold')
ax2.grid(True, alpha=0.3)

# Rolling VaR (95% confidence, 250-day window)
rolling_var_95 = portfolio_returns.rolling(window=250).apply(
    lambda x: np.percentile(x, 5) if len(x) == 250 else np.nan
)

ax3.plot(portfolio_returns.index, portfolio_returns, alpha=0.3, color='blue', label='Daily Returns')
ax3.plot(rolling_var_95.index, rolling_var_95, color='red', linewidth=2, label='Rolling VaR 95%')
ax3.set_title('Rolling VaR (95% Confidence, 250-day window)', fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)

# VaR exceedances (backtesting)
var_95 = var_results[0.95]
exceedances = portfolio_returns[portfolio_returns < var_95]
ax4.plot(portfolio_returns.index, portfolio_returns, alpha=0.5, color='blue', label='Daily Returns')
ax4.scatter(exceedances.index, exceedances, color='red', s=20, alpha=0.8, label='VaR Exceedances')
ax4.axhline(var_95, color='red', linestyle='--', label=f'VaR 95%: {var_95:.3f}')
ax4.set_title('VaR Backtesting: Exceedances', fontweight='bold')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
print("\n" + "="*50)
print("STEP 9: VaR BACKTESTING ANALYSIS")
print("="*50)

# Backtesting analysis
total_observations = len(portfolio_returns)

print("VaR Backtesting Results:")
print(f"Total Observations: {total_observations}")

for conf_level in confidence_levels:
    var_threshold = var_results[conf_level]
    exceedances = (portfolio_returns < var_threshold).sum()
    expected_exceedances = total_observations * (1 - conf_level)
    actual_rate = exceedances / total_observations
    
    print(f"\n{conf_level:.0%} Confidence Level:")
    print(f"  VaR Threshold: {var_threshold:.4f}")
    print(f"  Expected Exceedances: {expected_exceedances:.1f}")
    print(f"  Actual Exceedances: {exceedances}")
    print(f"  Expected Rate: {(1-conf_level):.1%}")
    print(f"  Actual Rate: {actual_rate:.1%}")
    
    # Simple traffic light system
    if abs(actual_rate - (1-conf_level)) < 0.01:
        status = "✅ GOOD"
    elif abs(actual_rate - (1-conf_level)) < 0.02:
        status = "⚠️  ACCEPTABLE"
    else:
        status = "❌ POOR"
    print(f"  Model Performance: {status}")

# STEP 10: PORTFOLIO VALUE VaR
print("\n" + "="*50)
print("STEP 10: PORTFOLIO VALUE AT RISK")
print("="*50)

# Assume a portfolio value for practical VaR calculation
portfolio_value = 1_000_000  # $1 million portfolio

print(f"Portfolio Value: ${portfolio_value:,.2f}")
print("\nValue at Risk (Dollar Amount):")

for conf_level in confidence_levels:
    var_dollar = abs(var_results[conf_level] * portfolio_value)
    es_dollar = abs(es_results[conf_level] * portfolio_value)
    
    print(f"\n{conf_level:.0%} Confidence Level:")
    print(f"  VaR (1-day): ${var_dollar:,.2f}")
    print(f"  Expected Shortfall: ${es_dollar:,.2f}")
    
    # Scaling to different time horizons (assuming i.i.d. returns)
    var_weekly = var_dollar * np.sqrt(5)
    var_monthly = var_dollar * np.sqrt(21)
    
    print(f"  VaR (1-week): ${var_weekly:,.2f}")
    print(f"  VaR (1-month): ${var_monthly:,.2f}")
```

```python
# STEP 11: COMPONENT VaR ANALYSIS
print("\n" + "="*50)
print("STEP 11: COMPONENT VaR ANALYSIS")
print("="*50)

# Calculate marginal VaR for each asset
print("Component VaR Analysis (95% Confidence Level):")

# Calculate portfolio volatility
portfolio_vol = portfolio_returns.std()
var_95_pct = var_results[0.95]

# Component VaR calculation
component_vars = {}
marginal_vars = {}

for i, asset in enumerate(assets):
    # Marginal VaR: sensitivity of portfolio VaR to small change in weight
    asset_portfolio_cov = returns[asset].cov(portfolio_returns)
    marginal_var = asset_portfolio_cov / portfolio_vol * var_95_pct / portfolio_vol
    marginal_vars[asset] = marginal_var
    
    # Component VaR: contribution of each asset to total VaR
    component_var = weights[i] * marginal_var
    component_vars[asset] = component_var
    
    print(f"\n{asset}:")
    print(f"  Weight: {weights[i]:.1%}")
    print(f"  Marginal VaR: {marginal_var:.6f}")
    print(f"  Component VaR: {component_var:.6f}")
    print(f"  Component VaR ($): ${abs(component_var * portfolio_value):,.2f}")

# Verify that component VaRs sum to total VaR
total_component_var = sum(component_vars.values())
print(f"\nVerification:")
print(f"Sum of Component VaRs: {total_component_var:.6f}")
print(f"Portfolio VaR (95%): {var_95_pct:.6f}")
print(f"Difference: {abs(total_component_var - var_95_pct):.8f}")

```

```python
# STEP 12: FINAL SUMMARY VISUALIZATION
print("\n" + "="*50)
print("STEP 12: FINAL SUMMARY VISUALIZATION")
print("="*50)

# Create a summary dashboard
fig = plt.figure(figsize=(20, 16))

# Create a 3x3 grid
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# 1. Portfolio composition
ax1 = fig.add_subplot(gs[0, 0])
ax1.pie(weights, labels=assets, autopct='%1.1f%%', startangle=90)
ax1.set_title('Portfolio Composition', fontweight='bold')

# 2. VaR comparison chart
ax2 = fig.add_subplot(gs[0, 1])
conf_labels = [f"{int(c*100)}%" for c in confidence_levels]
var_values = [abs(var_results[c]) for c in confidence_levels]
es_values = [abs(es_results[c]) for c in confidence_levels]

x = np.arange(len(conf_labels))
width = 0.35

ax2.bar(x - width/2, var_values, width, label='VaR', color='lightcoral')
ax2.bar(x + width/2, es_values, width, label='Expected Shortfall', color='darkred')
ax2.set_xlabel('Confidence Level')
ax2.set_ylabel('Loss Rate')
ax2.set_title('VaR vs Expected Shortfall', fontweight='bold')
ax2.set_xticks(x)
ax2.set_xticklabels(conf_labels)
ax2.legend()
ax2.grid(True, alpha=0.3)

# 3. Component VaR
ax3 = fig.add_subplot(gs[0, 2])
comp_var_values = [abs(component_vars[asset] * portfolio_value) for asset in assets]
colors = plt.cm.Set3(np.linspace(0, 1, len(assets)))
bars = ax3.bar(assets, comp_var_values, color=colors)
ax3.set_title('Component VaR by Asset ($)', fontweight='bold')
ax3.set_ylabel('Dollar Amount')
for bar, value in zip(bars, comp_var_values):
    ax3.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(comp_var_values)*0.01,
             f'${value:,.0f}', ha='center', va='bottom', fontsize=8)

# 4. Returns distribution with VaR
ax4 = fig.add_subplot(gs[1, :])
portfolio_returns.hist(bins=100, ax=ax4, alpha=0.7, color='lightblue', edgecolor='black')
for i, conf_level in enumerate(confidence_levels):
    ax4.axvline(var_results[conf_level], color=colors[i], linestyle='--', linewidth=2,
               label=f'VaR {conf_level:.0%}: {var_results[conf_level]:.3f}')
ax4.set_title('Portfolio Returns Distribution with VaR Levels', fontweight='bold', fontsize=14)
ax4.legend()
ax4.grid(True, alpha=0.3)

# 5. Risk metrics table
ax5 = fig.add_subplot(gs[2, :])
ax5.axis('tight')
ax5.axis('off')

# Create summary table data
summary_data = []
summary_data.append(['Metric', 'Value'])
summary_data.append(['Portfolio Value', f'${portfolio_value:,.2f}'])
summary_data.append(['Mean Daily Return', f'{portfolio_returns.mean():.4f} ({portfolio_returns.mean()*252:.2%} ann.)'])
summary_data.append(['Daily Volatility', f'{portfolio_returns.std():.4f} ({portfolio_returns.std()*np.sqrt(252):.2%} ann.)'])
summary_data.append(['Sharpe Ratio', f'{portfolio_returns.mean()/portfolio_returns.std()*np.sqrt(252):.2f}'])

for conf_level in confidence_levels:
    var_dollar = abs(var_results[conf_level] * portfolio_value)
    es_dollar = abs(es_results[conf_level] * portfolio_value)
    summary_data.append([f'VaR {conf_level:.0%} (1-day)', f'${var_dollar:,.2f} ({var_results[conf_level]:.2%})'])
    summary_data.append([f'Expected Shortfall {conf_level:.0%}', f'${es_dollar:,.2f} ({es_results[conf_level]:.2%})'])

table = ax5.table(cellText=summary_data[1:], colLabels=summary_data[0], 
                 cellLoc='center', loc='center', bbox=[0, 0, 1, 1])
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 2)

# Style the table
for (i, j), cell in table.get_celld().items():
    if i == 0:
        cell.set_text_props(weight='bold')
        cell.set_facecolor('#40466e')
        cell.set_text_props(color='white')
    else:
        cell.set_facecolor('#f5f5f5' if i % 2 == 0 else 'white')

ax5.set_title('Risk Metrics Summary', fontweight='bold', fontsize=14, pad=20)

plt.suptitle('Historical VaR Analysis - Portfolio Risk Dashboard', 
             fontsize=18, fontweight='bold', y=0.98)
plt.savefig("p92.png",dpi=300)
plt.show()

```
