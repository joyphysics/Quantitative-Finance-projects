### What is "Adjusted Close" in Stocks?

Adjusted Close (or Adjusted Closing Price) is the stock's closing price adjusted for corporate actions like:
- Dividends
- Stock splits
- Reverse splits
- Spin-offs

It represents the real economic value of the stock at market close, taking into account anything that changes the effective value per share.

#### Example

Let's say a stock closed at $\$100$ yesterday, and today it pays a $\$2$ dividend.
- Raw Close today: $\$98$ (price drops by the dividend)
- Adjusted Close today: $\$100$ (reflects "true" value continuity)

This helps you compare historical prices fairly over time.

- Use close if you care about exact market prices on a given day.
- Use Adjusted Close if you care about investment returns over time.

```python
adj_close_prices = df['Adj Close']
adj_close_prices.head()
```

### Daily Return Formula

The daily return measures how much a stock price changes from one day to the next, usually based on adjusted close prices.

Formula:

\begin{equation*}
\text { Daily Return }_t=\frac{P_t-P_{t-1}}{P_{t-1}}=\frac{P_t}{P_{t-1}}-1
\end{equation*}


Where:
- $P_t=$ price at day $t$
- $P_{t-1}=$ price at previous day

This gives the percentage change, e.g., 0.02 means a $2 \%$ gain.

```python
daily_returns = adj_close_prices.pct_change()
print(daily_returns.head())
```

```python
daily_returns = (adj_close_prices / adj_close_prices.shift(1)) - 1
print(daily_returns.head())
```

### What is Log Daily Return?

The log return (also called continuously compounded return) is another way to measure daily return, preferred in many financial models because of its nice mathematical properties.

Log Return Formula

\begin{equation*}
\log \text { Return }_t=\ln \left(\frac{P_t}{P_{t-1}}\right)
\end{equation*}


Where:
- $P_t=$ adjusted closing price at time $t$
- ln = natural logarithm
- 
| Feature | Benefit |
| :--- | :--- |
| Additive over time | Total return over $n$ days = sum of log returns |
| Symmetric handling | Downward and upward moves are proportionally fair |
| Used in modeling | Works better in many mathematical models (e.g., Black-Scholes) |

```python
## Log return
log_returns = np.log(adj_close_prices / adj_close_prices.shift(1))
print(log_returns.head())
```

### Cumulative Log Return

Given daily log returns:

\begin{equation*}
\log _{\text {Return }}^t=\ln \left(\frac{P_t}{P_{t-1}}\right)
\end{equation*}


The cumulative log return over time is the sum of log returns up to time $t$ :

\begin{equation*}
\text { Cumulative Log Return }{ }_t=\sum_{i=1}^t \ln \left(\frac{P_i}{P_{i-1}}\right)
\end{equation*}


Then, convert this to cumulative simple return (percentage gain from start):

\begin{equation*}
\text { Cumulative Return }_t=e^{\text {Cumulative Log Return }_t}-1
\end{equation*}


```python
# Cumulative log returns (running sum)
cumulative_log_returns = log_returns.cumsum()

# Convert to cumulative simple returns
cumulative_returns = np.exp(cumulative_log_returns) - 1

print(cumulative_returns.head())
```

```python
cumulative_returns.plot(figsize=(10, 5))
plt.title("Cumulative Return (from Returns)")
plt.ylabel("Total Return (%)")
plt.xlabel("Date")
plt.grid(True)
plt.savefig("cumulative_simple_return.png", dpi=300)
plt.show()
```

```python
cumulative_log_returns.plot(figsize=(10, 5))
plt.title("Cumulative Return (from Log Returns)")
plt.ylabel("Total Return (%)")
plt.xlabel("Date")
plt.grid(True)
plt.savefig("cumulative_log_return.png", dpi=300)
plt.show()
```

### Export data

```python
export_df = pd.concat([
    adj_close_prices.add_suffix(" - Adj Close"),
    log_returns.add_suffix(" - Log Return"),
    cumulative_returns.add_suffix(" - Cumulative Return")
], axis=1)

# Step 4: Save to CSV
export_df.to_csv("financial_data_export.csv")

print(" Data exported to 'financial_data_export.csv'")
```
