# Jump Diffusion

### 1. Introduction

The Jump Diffusion Model is a stochastic model widely used in quantitative finance to describe asset prices exhibiting both continuous price fluctuations (diffusions) and sudden, discrete price jumps. The most common variant is the Merton Jump-Diffusion Model, developed by Robert Merton in 1976. It extends the classical Geometric Brownian Motion (GBM) used in the Black-Scholes model by incorporating rare but significant jumps.

### 2. Why Jump Diffusion?

Traditional models, such as Black-Scholes (based purely on continuous processes), fail to capture significant empirical features:
- Fat-tailed distributions: Observed price returns often have heavier tails than those predicted by the normal distribution.
- Implied volatility smiles and skews: Option markets frequently exhibit volatility patterns inconsistent with standard GBM.
- Sudden market moves: Large and abrupt price changes due to unexpected news/events.

The Jump Diffusion Model addresses these shortcomings.


### 3. Mathematical Formulation

Merton's Jump Diffusion Model:
The stock price $S_t$ evolves according to the stochastic differential equation (SDE):

\begin{equation*}
\frac{d S_t}{S_t}=\mu d t+\sigma d W_t+d J_t
\end{equation*}

where:
- $\mu$ is the drift (expected rate of return).
- $\sigma$ is the volatility (standard deviation of continuous returns).
- $W_t$ is a standard Brownian motion (Wiener process).
- $J_t$ is a pure jump process independent of $W_t$.

Jump Component $J_t$ :
- $J_t$ represents the cumulative jump size at time $t$.
- It is modeled typically as a compound Poisson process:

\begin{equation*}
J_t=\sum_{i=1}^{N_t}\left(Y_i-1\right)
\end{equation*}

or equivalently:

\begin{equation*}
S_t=S_0 \exp \left[\left(\mu-\frac{1}{2} \sigma^2-\lambda \kappa\right) t+\sigma W_t\right] \prod_{i=1}^{N_t} Y_i
\end{equation*}


where:
- $\quad N_t$ is a Poisson counting process with jump intensity $\lambda$, i.e., the probability of jumps per unit time is Poisson-distributed:

\begin{equation*}
N_t \sim \text { Poisson }(\lambda t)
\end{equation*}

- The jump magnitudes $Y_i$ are independent and identically distributed (IID) random variables, typically lognormal:

\begin{equation*}
Y_i \sim \operatorname{Lognormal}\left(\alpha, \delta^2\right)
\end{equation*}

- Thus, each jump represents a multiplicative jump factor from $S \rightarrow Y_i S$.

- $\kappa=\mathbb{E}[Y-1]$ : Mean percentage jump

####  Brownian Motion and GBM Recap

Under the Black-Scholes/GBM framework:

\begin{equation*}
d S_t=\mu S_t d t+\sigma S_t d W_t, \quad S_0>0
\end{equation*}

where:
- $W_t$ is a standard Brownian motion under the real-world measure $\mathbb{P}$,
- $\mu$ is the drift,
- $\sigma>0$ is the volatility.

The logarithm of $S_t$ satisfies

\begin{equation*}
d\left(\ln S_t\right)=\left(\mu-\frac{1}{2} \sigma^2\right) d t+\sigma d W_t
\end{equation*}

so that

\begin{equation*}
S_t=S_0 \exp \left[\left(\mu-\frac{1}{2} \sigma^2\right) t+\sigma W_t\right]
\end{equation*}


Under risk-neutral valuation, one replaces $\mu$ by $r$, the risk-free rate.

#### Poisson Process

A Poisson process $N_t$ with intensity $\lambda>0$ is defined by:
- $N_0=0$
- For $0 \leq t_1<t_2<\cdots<t_n$, the increments $N_{t_j}-N_{t_{j-1}}$ are independent and each has a Poisson distribution with parameter $\lambda\left(t_j-t_{j-1}\right)$.
- In particular,

\begin{equation*}
\operatorname{Pr}\left\{N_t=n\right\}=\exp (-\lambda t) \frac{(\lambda t)^n}{n!}, \quad n=0,1,2, \ldots
\end{equation*}

- The expected number of jumps up to time $t$ is $\mathbb{E}\left[N_t\right]=\lambda t$.

####  Jump-Size Distribution

Whenever a jump occurs (i.e., at each arrival time of $N_t$ ), the asset price multiplies by a random factor $Y$. We typically assume:

\begin{equation*}
\ln Y \sim \mathcal{N}\left(\alpha, \delta^2\right), \quad Y>0
\end{equation*}

so that $Y$ is lognormally distributed with

\begin{equation*}
\mathbb{E}[Y]=\exp \left(\alpha+\frac{1}{2} \delta^2\right), \quad \operatorname{Var}(Y)=\left(e^{\delta^2}-1\right) e^{2 \alpha+\delta^2}
\end{equation*}


Equivalently, one can denote the jump size increment as $J:=Y-1$, so that at each jump time the price increases from $S_{t^{-}}$to $S_{t^{+}}=S_{t^{-}}(1+J)$.

### Constructing the Jump Diffusion SDE

We now combine the continuous GBM piece with the jump piece. The Merton Jump-Diffusion SDE for $S_t$ is:

\begin{equation*}
\frac{d S_t}{S_{t^{-}}}=\mu d t+\sigma d W_t+(Y-1) d N_t
\end{equation*}

where:
- $S_t$ - denotes the left-limit of $S$ just before time $t$.
- $\mu d t+\sigma d W_t$ is the continuous (diffusion) part.
- ( $Y-1$ ) $d N_t$ is the jump part: if $d N_t=1$ (one jump occurs in $[t, t+d t$ )), then $S$ is multiplied by $Y$. If $d N_t=0$, there is no jump in that small interval.
3.1 Interpretation
- When $d N_t=0$, the SDE reduces to $d S_t=\mu S_t d t+\sigma S_t d W_t$.
- When $d N_t=1$, simultaneously with the Brownian increment, the stock price jumps:

\begin{equation*}
S_{t^{+}}=S_{t^{-}}+S_{t^{-}}(Y-1)=S_{t^{-}} Y .
\end{equation*}

- The process $N_t$ and the i.i.d. jump magnitudes $\left\{Y_i\right\}$ are assumed independent of the Brownian motion $W_t$.

### Explicit Integral Solution

To see how $S_t$ evolves over time, it is often easier to work with the log-price:

\begin{equation*}
X_t:=\ln S_t .
\end{equation*}


Applying It√¥'s lemma for jump processes, one obtains:

\begin{equation*}
d X_t=\ln \left(\frac{S_{t^{+}}}{S_{t^{-}}}\right)+\left(\mu-\frac{1}{2} \sigma^2\right) d t+\sigma d W_t
\end{equation*}

but $\ln \left(\frac{S_{t^{+}}}{S_{t^{-}}}\right)=\ln Y$ if a jump occurs, and 0 if no jump. More formally,

\begin{equation*}
d X_t=\left(\mu-\frac{1}{2} \sigma^2\right) d t+\sigma d W_t+\ln (Y) d N_t .
\end{equation*}


Integrating from 0 to $t$ :

\begin{equation*}
X_t=\ln S_0+\left(\mu-\frac{1}{2} \sigma^2\right) t+\sigma W_t+\sum_{i=1}^{N_t} \ln Y_i .
\end{equation*}


Hence

\begin{equation*}
S_t=S_0 \exp \left[\left(\mu-\frac{1}{2} \sigma^2\right) t+\sigma W_t\right] \times \prod_{i=1}^{N_t} Y_i
\end{equation*}


Equivalently, writing $J_i:=Y_i-1$,

\begin{equation*}
S_t=S_0 \exp \left[\left(\mu-\frac{1}{2} \sigma^2\right) t+\sigma W_t\right] \prod_{i=1}^{N_t}\left(1+J_i\right) .
\end{equation*}


### 4. Step-by-Step Derivation

##### Step 1: Model Setup
We start with the logarithmic price process. Define $X_t=\ln \left(S_t\right)$ :

\begin{equation*}
d X_t=\left(\mu-\frac{1}{2} \sigma^2\right) d t+\sigma d W_t+d \ln \left(J_t+1\right)
\end{equation*}


Note: We add 1 to $J_t$ explicitly to denote proportional jumps clearly.

##### Step 2: Jump Dynamics
- Consider the jump process over a small interval $d t$ :
- Probability of exactly one jump is $\lambda d t$.
- Probability of more than one jump in $d t$ is $o(d t)$ (negligible).
- Probability of no jump is $1-\lambda d t$.

Thus, the jump part at time $t$ has increments:

\begin{equation*}
d J_t= \begin{cases}Y-1, & \text { with probability } \lambda d t \\ 0, & \text { with probability } 1-\lambda d t\end{cases}
\end{equation*}

##### Step 3: Expectations and Variance of Jumps
Define the random variable for a jump magnitude as $Y$, and assume it follows a lognormal distribution. Then we have:
- Mean jump size:

\begin{equation*}
\mathbb{E}[Y]=e^{\alpha+\frac{\delta^2}{2}}
\end{equation*}

- Variance of jump size:

\begin{equation*}
\operatorname{Var}(Y)=\left(e^{\delta^2}-1\right) e^{2 \alpha+\delta^2}
\end{equation*}


##### Step 4: Risk-Neutral Dynamics and Drift Correction
For derivative pricing, we work under the risk-neutral measure $Q$. Under $Q$, discounted asset prices must be martingales:
- The discounted price $e^{-r t} S_t$ must have no drift under $Q$. Hence, we adjust drift from $\mu$ to $r-$ $\lambda(\mathbb{E}[Y]-1):$

Thus, under the risk-neutral measure:

\begin{equation*}
\frac{d S_t}{S_t}=(r-\lambda \kappa) d t+\sigma d W_t^Q+d J_t
\end{equation*}

where:
- $r$ is the risk-free rate.
- $W_t^Q$ is Brownian motion under the risk-neutral measure.
- $\kappa=\mathbb{E}[Y-1]$ is the mean jump size increment.

This ensures the discounted process has zero expected drift under $Q$.


### 5. Analytical Pricing under Merton's Model

Merton provided a closed-form solution for European call options:

\begin{equation*}
C(S, K, T)=\sum_{n=0}^{\infty} e^{-\lambda^{\prime} T} \frac{\left(\lambda^{\prime} T\right)^n}{n!} C_{B S}\left(S, K, T ; r_n, \sigma_n\right)
\end{equation*}

where:
- $\lambda^{\prime}=\lambda \mathbb{E}[Y]$, adjusted jump intensity.
- $C_{B S}$ is the classical Black-Scholes formula.
- Adjusted volatility and drift are:
- $\sigma_n^2=\sigma^2+\frac{n \delta^2}{T}$
- $r_n=r-\lambda \kappa+\frac{n\left(\alpha+\frac{\delta^2}{2}\right)}{T}$

This pricing formula is essentially a Poisson-weighted average of classical Black-Scholes prices.

### 6. Applications and Limitations

##### Applications:
- Pricing options and derivatives exhibiting jumps.
- Risk management and Value-at-Risk (VaR) calculations.
- Hedging strategies that account for sudden market moves.

##### Limitations:
- More parameters ( $\lambda, \alpha, \delta$ ) add complexity to estimation.
- Requires calibration against market data (option prices or historical jump events).
- Assumes jumps are IID, which might not fully represent correlated or clustered jumps.
